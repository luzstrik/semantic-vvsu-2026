```python
import requests
URL = "http://localhost:8000/generate"
```

```python
tests = [
    "Рассчитай КБЖУ для мужчины 82 кг, 178 см, 27 лет, цель — похудение.",
    "Сделай программу тренировок на 3 дня для новичка в зале на набор мышечной массы.",
    
    "Сколько калорий и БЖУ нужно девушке 60 кг, 165 см, 24 года для поддержания веса?",
    "Подбери workout-программу для похудения с упором на кардио и ноги.",
   
    "Как дела?",
    "Расскажи анекдот про фитнесс",
]
```

```python
import requests
import json

URL = "http://127.0.0.1:8000/generate"


def generator(tests):
    for message in tests:
        response = requests.post(
            URL,
            json={"message": message}
        )
        data = response.json()
        response_type = data.get("type")
        answer = data.get("answer")
        macros = data.get("macros")
        workout = data.get("workout")
        formatted_response = {
            "type": response_type,
            "answer": answer,
            "macros": macros,
            "workout": workout,
        }
        yield (
            f"\nQuestion:\n{message}\n\n"
            f"{'=' * 120}\n\n"
            f"Response:\n"
            f"{json.dumps(formatted_response, ensure_ascii=False, indent=2)}\n"
        )

gen = generator(tests)
```
