```python
import requests
URL = "http://127.0.0.1:8000/generate"
```

```python

tests = [
    "как дела?",
    "расскажи анекдот",

    "Я вешу 80 кг, рост 180, возраст 25, хочу похудеть. Посчитай КБЖУ",
    "Рассчитай КБЖУ для мужчины 90 кг вес, рост - 185 см, возраст - 30 лет, цель набор массы",

    "Составь программу тренировок на неделю",
    "Сделай план тренировок дома без оборудования",
]
```

```python
def generator(tests):
    for message in tests:
        response = requests.post(URL, json={"message": message})
        data = response.json()
        answer = data.get("answer", "")

        yield f"Question: {message}\n{'*' * 120}\nResponse: {answer}"


gen = generator(tests)
```
